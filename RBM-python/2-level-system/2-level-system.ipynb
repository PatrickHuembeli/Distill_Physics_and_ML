```python
import numpy as np
import matplotlib.pyplot as plt

temp = np.linspace(0.001, 10, 100)
epsilon = 0
delta = 1.0

def prob_excited(temp, delta):
    return 1/(1+np.exp(delta/temp))

def prob_ground(temp, delta):
    return 1/(1+np.exp(-delta/temp))

occup_e = prob_excited(temp, delta)
occup_g = prob_ground(temp, delta)
```

```python
plt.plot(occup_e)
plt.plot(occup_g)
plt.show()
```
