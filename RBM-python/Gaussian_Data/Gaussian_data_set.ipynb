This file is to generate a dataset for a simple RBM training

```
import numpy as np
import matplotlib.pyplot as plt

number_of_samples = 10000
mu = 0.0
sigma = 1.0
x_min = -4.0
x_max = 4.0
x_axis = np.linspace(x_min, x_max, 101)

def gauss(x, mu, sigma):
  return 1/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(mu - x)**2/(2*sigma**2))

g = gauss(x_axis, mu, sigma)

plt.plot(x_axis, g)
plt.show()

dataset = []
a = 0
for j in range(number_of_samples):
  sample = np.random.binomial(1, g)
  dataset.append(sample)
  a += sample
  
plt.plot(a)
plt.show()
```
